# Outlier Remover

```python
figure = plt.figure(figsize = (8,10))
for i,j in enumerate(num.columns):
  plt.subplot(3,3,i+1)
  sns.boxplot(num[j])
  plt.title(j)
 
```

```python
z.head()
```

```python
z.size
```

```python
Q1 = z.quantile(0.25)
Q3 = z.quantile(0.75)
IQR = Q3 - Q1
z = z[~((z < (Q1 - 1.5 * IQR)) |(z > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
z.size
```

```python
z.head()
```
